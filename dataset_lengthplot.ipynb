```python
from datasets import load_from_disk

dataset_train = load_from_disk('/home/kkj/axolotl/datasets/IPR036736_90_grouped/train')
dataset_test = load_from_disk('/home/kkj/axolotl/datasets/IPR036736_90_grouped/test')
dataset = dataset_train.with_format('torch')

len(dataset_train), len(dataset_test)
```

```python
# Add this code to a new cell in your notebook

import matplotlib.pyplot as plt
import numpy as np

# Extract the 'length' column
lengths = dataset['length']


# Calculate the median
median_lengths = [min(np.median(length), 4096) for length in lengths]

# Plot the data
plt.figure(figsize=(8, 6)),
plt.hist(median_lengths, bins=32, color='skyblue', edgecolor='black', alpha=0.7),
plt.title('Median cluster sequence lengths'),
plt.xlabel('Length'),
plt.ylabel('Amount'),
plt.legend(),
plt.grid(axis='y', linestyle='--', alpha=0.7),
plt.show()
```
